```python
import gensim
from gensim.models import Word2Vec,KeyedVectors
```

```python
%%time

# path to dataset
word2vec_path = "/kaggle/input/googlenewsvectorsnegative300/GoogleNews-vectors-negative300.bin"

# Load 500,000 most common words
model = gensim.models.KeyedVectors.load_word2vec_format(word2vec_path, binary = True, limit = 500000)
```

```python
model['cricket']
```

```python
model.most_similar('man')
```

```python
model.most_similar('cricket')
```

```python
model.most_similar('facebook')
```

```python
model.similarity('man','woman')
```

```python
model.similarity('man','PHP')
```

```python
model.doesnt_match(['PHP','java','monkey'])
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```

```python
vec = model['INR'] - model ['India'] + model['England']
model.most_similar([vec])
```
